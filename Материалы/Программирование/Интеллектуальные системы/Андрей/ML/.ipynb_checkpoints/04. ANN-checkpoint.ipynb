Библиотека Keras
----------------

Библиотека Keras — популярное средство моделирования искусственных нейронных сетей, позволяющее быстро создавать эффективные модели машинного обучения. Для вычислений она использует вспомогатеьные библиотеки (backends): Theano (Монреальский университет) или Tensorflow (Google).

Основной вид нейросетевых моделей в Keras — последовательные (Sequential). В таких моделях слои располагаются последовательно. Сами слои (layers) либо перечисляются в виде списка в конструкторе, либо добавляются по очереди с помощью метода `add`.

Построим нейронную сеть для классифиации ирисов Фишера. Загрузим данные:

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris['data']
y = iris['target']

m, n = X.shape
print(X.shape)
```

В `y` содержатся номера классов. Keras требует, чтобы в задачах классификации класс представлялся в виде вектора, в котором все значения нулевые, а в позиции, соответствующей классу стояла бы единица. Например, для трёхклассовой задачи класс 0 кодировался бы как $(1, 0, 0)$, а класс 2 как $(0, 0, 1)$. Такое преобразование легко сделать средствами Keras:

```python
from keras.utils.np_utils import to_categorical
yc = to_categorical(y)
print(y)
```

Разобъём выборку на обучающую и тестовую:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, yc_train, yc_test = train_test_split(X, yc, test_size=0.25)
```

Есть несколько видов слоёв:
- `Dense` — полносвязный (каждый с каждым) слой без функции активации,
- `Activation` — слой функций активации (например, использующих сигмоиду),
- `Dropout` - особый слой, который позволяет исключить часть связей для ускорения обучения и снижения переобученности
- и многие другие.

Рассмотренных выше типов слоёв достаточно для построения достаточно эффективных сетей.

Каждый слой при создании может принимать дополнительные параметры. Для `Dense` это число нейронов и количество входов (`input_dim`). Для `Dropout` — доля случайным образом выбираемых элементов, которые обнуляются на одном шаге обучения. У `Activation` — вид функции активации.

Распространённые функции активации:
- `sigmoid` — классическая сигмоидальная функция $\frac1{1+exp(-z)}$;
- `tanh` — гиперболический тангенс $\mathrm{th}\,z$, область значений равна $[-1, 1]$;
- `relu` — т. н. «выпрямитель» (rectifier) $ReLU(x) = \max(0, x)$,
- `softmax` — мягкий максимум, аналог обычного максимума, нормированный на единицу,
- и другие.

В настоящее время чаще всего используется функция ReLU. Она кусочно-линейная, поэтому вычисляется намного быстрее, что хорошо сказывается на скорости обучения. С другой стороны у неё есть горизонтальная область, на которой градиент равен нулю, что может негативно сказаться на точности обучения.

Пусть сеть имеет $n=4$ входа (по числу признаков), $64$ нейрона в скрытом слое и $3$ выхода (по одному для каждого класса). Будем обозначать такую архитектуру 4—64—3. После скрытого слоя при обучени с вероятностью 0,5 будет происходить исключение нейронов из модели для устранения переобученности.

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout

model = Sequential()
model.add(Dense(64, input_shape=(n, )))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(3))
model.add(Activation('softmax'))
```

В последнем слое использовалась функция `softmax`, так как у сети по выходу для каждого из классов. И  нужно выбирать тот класс, у которого 

Выведем информацию о модели.

```python
model.summary()
```

После того, как модель создана, её необходимо скомпилировать. При этом компиляция происходит в прямом смысле: для повышения быстродействия библиотека генерирует программу на языке C, которая затем компилируется в машинные команды, эффективно выполяемые компьютером. В зависимости от настроек может генерироваться как код для процессора (CPU), так и для видеокарты (GPU). Последнее предпочтительнее из-за значительного ускорения вычислений.

При компиляции указываются следующие параметры:
- тип оптимизатора (`optimizer`) — метод оптимизации, используемый для минимизации функционала качества. Например:
  - `sgd` — стохастический градиентный спуск,
  - `rmsprop`,
  - `adagrad`,
  - `adadelta`,
  - `adam` и другие.
- функционал качества (`loss`):
  - `mse` — среднеквадратическая ошибка (для числовых ответов),
  - `mae` — средняя абсолютная ошибка (для числовых ответов),
  - `binary_crossentropy` — бинарная кроссэнтропия (для бинарной классификации),
  - `categorical_crossentropy` — категориальная кроссэнтропия (для многоклассовой классфикации),
  - `cosine_proximity` — средний косинус угла между предсказанным и истинным векторами
  - и другие.
  
Также можно указать метрику (`metrics`), вычисляемую во время обучения. Она не влияет на само обучение, но позволяет следить за его точностью.

Задача классификации ирисов — многоклассовая, поэтому компиляция выполняется следующим образом:

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
print(yc_train.shape)
```

Модели Keras имеют интерфейс, похожий на модели Scikit-Learn. Обучение модели так же выполняется методом `fit`. Дополнительно можно указать:
- nb_epoch — количество эпох обучения,
- batch_size — размер блока прецедентов для обучения на одном этапе; обучение выполняется блоками а не сразу на всех данных для ускорения расчётов.

```python
model.fit(X_train, yc_train, nb_epoch=100, batch_size=32)
```

Оценим точность модели на тестовой выборке:

```python
score = model.evaluate(X_test, yc_test, verbose=0)
print('Потери:', score[0])
print('Точность обучения:', score[1])
```

Применять модель к данным можно с помощью метода `predict`.

```python
import numpy as np
model.predict(np.array([[ 4.9,  3. ,  1.4,  0.2]]))
```

Наибольшее значение у первого компонента вектора, следовательно ответ — «класс 0» (это верный ответ).

База MNIST
----------

База изображений рукописных цифр MNIST — открытая база, созданная Национальным институтом стандартизации и технологий (NIST) США. База содержит 60000 изображений 28×28 пикселей (784 признака), которые используются для обучения, и 10000 — для тестирования. Она широко используется для тестированияалгоритмов обучения, поэтому в Keras есть встроенная функция для загрузки её из сети интернет.

```python
from keras.datasets import mnist
import numpy as np
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

Выведем случайное изображение и номер класса.

```python
import random
import matplotlib.pyplot as plt
%matplotlib inline

m_test, _, _ = X_test.shape 

k = random.randint(0, m_test)
Xrand = X_test[k]
yrand = y_test[k]

plt.imshow(Xrand, cmap='gray')
plt.show()

print('Класс:', yrand)
```

Для обучения нейронной сети данные в представленном виде не очень подходят. Поэтому нормируем все значения на 1 и представим изображения (28×28) в виде вектора (784). Также номера классов нужно преобразовать в матрицу.

```python
# Выписываем в виде вектора
X_train_n = X_train.reshape(60000, 784)
X_test_n  = X_test.reshape(10000, 784)

# Нормируем на 1
X_train_n = X_train_n.astype('float32')
X_test_n  = X_test_n.astype('float32')
X_train_n /= 255
X_test_n  /= 255

# Преобразуем номера классов в матрицу классов
y_train_n = to_categorical(y_train, 10)
y_test_n  = to_categorical(y_test, 10)
n, p = X_train_n.shape
print(y_train_n.shape)
```

Используйте для обучения данные, хранящиеся именно в переменных с суффиксом «_n».

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout

model = Sequential()
model.add(Dense(128, input_shape=(784, )))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(128,))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(10))
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
print(y_test_n.shape)
```

```python
model.fit(X_train_n, y_train_n, nb_epoch=5, batch_size=128)
```

```python
score = model.evaluate(X_test_n, y_test_n, verbose=0)
print('Потери:', score[0])
print('Точность обучения:', score[1])
```

2. Выводим на экран случайное изображение из базы MNIST, правильный ответ и предсказанное сетью значение класса

```python
import random
import matplotlib.pyplot as plt
%matplotlib inline

m_test, _, _ = X_test.shape 
kol = 0
k = random.randint(0, m_test)
for i in range(0, 10000):
    yrand = y_test[i]
    Xrand = X_test[i]
    y_pred = np.argmax(model.predict(Xrand.reshape(1, 784)))
    if yrand != y_pred:
        kol += 1 
        if kol < 20 :
            plt.imshow(Xrand, cmap='gray')
            plt.show()
            print('Класс:', yrand)
            print('Предсказано:', y_pred)
print("Кол-во ошибок", kol)
```
